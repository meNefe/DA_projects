# Исследование факторов, влияющих на популярность видеоигр.

**Цель исследования:** 
- определить характеристики популярных игр и основные предпочтения игроков, в том числе по регионам;
- оценить влияние различных факторов на продажи игр;
- наметить ориентиры для закупки игр на прогнозируемый период.


Ничего не известно о качестве данных, поэтому для начала необходимо оценить их полноту и исправить ошибки.
Перед нами исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы . Выявим определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании интернет-магазина.


Описание данных:
 - Name — название игры
 - Platform — платформа
 - Year_of_Release — год выпуска
 - Genre — жанр игры
 - NA_sales — продажи в Северной Америке (миллионы проданных копий)
 - EU_sales — продажи в Европе (миллионы проданных копий)
 - JP_sales — продажи в Японии (миллионы проданных копий)
 - Other_sales — продажи в других странах (миллионы проданных копий)
 - Critic_Score — оценка критиков (максимум 100)
 - User_Score — оценка пользователей (максимум 10)
 - Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

**Анализ будет включать в себя следующие этапы:**
- Загрузка и знакомство с данными
- Предобработка
- Расчет глобальных продаж
- Исследование данных по динамике продаж, жизненном цикле платформ, различиях в жанрах
- Описание портрета пользователя по регионам
- Оценка гипотез по влиянию жанра и платформы на рейтинг

##### Жизненный цикл платформы.

Посмотрим на распределение выпуска игр по годам на основании топ-10 платформ по количеству продаж игр.

```python
for platform, group_data in df.groupby('platform'):
    if platform in platforms.index:
        group_data.plot(
            y = 'year_of_release',
            kind = 'hist',
            bins = 50,
            title = 'Выпуск игр на {platform}.'.format(platform = platform),
            color = '#2e8f87',
            figsize = (8, 6),
            legend = False)
        plt.ylabel('Кол-во игр')
        plt.xlabel('Год');
```

Прежде чем оценивать жизненный цикл, избавимся от ошибки, которая имеется в выборке по платформе DS (Nintendo DS появилась в 2004 году, соответственно игр в 1985 г. на нее не выпускалось).

```python
df.drop(df.loc[(df['year_of_release'] == 1985)&(df['platform'] == 'DS')].index, inplace=True)
df.loc[df['platform']  == 'DS', 'year_of_release'].unique()
```

Как мы видим, есть определенный период, за который обычно консоль морально устаревала и на ее смену выпускались новые усовершенствованные модели. Этот цикл примерно равен 5-6 годам. Есть долгоиграющие платформы, как xbox360 и PlayStation3 - у них цикл длится около 10 лет. Для ПК естественно мы не наблюдаем такой картины. Продажи игр на ПК подвержены циклам спада и подъема, но с учетом постепенного безпрерывного усовершенствования железа и разработке доп.гаджетов спрос на игры для ПК будет всегда.

```python
temp = df.copy() 
time_life = temp.pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max']).reset_index()
time_life['life'] = time_life['max'] - time_life['min'] + 1 # в срок жизни платформы, должны входить границы диапазона, 
                                                            # поэтому +1
time_life['life'].median()
```

На основании полученных диаграмм и расчетов получаем 7 лет жизненного цикла платформы (т.е. включая период роста и падения). Возьмем для дальнейшего анализа 4 года (включая неполный 2016г.) т.к. это медиана жизненного цикла платформы.

Посчитаем коэфициенты корреляции, общие для всех платформ.

```python
relations = ['total_sales', 'critic_score', 'user_score']
plt.figure(figsize=(7,6))
sns.heatmap(df[relations].corr(), annot=True, cmap='twilight')
plt.title('Зависимость глобальных продаж (млн. копий) PS4 и рейтинга игры')
plt.ylabel('Параметры модели')
plt.xlabel('Параметры модели')
plt.show()
```

И здесь расчеты не выдают прямой корреляции -даже связь между пользовательским рейтингом и рейтингом критиков имеет большую корреляцию между собой.

Посчитаем корреляцию продаж и рейтинга для отдельных платформ.

```python
fig, ax = plt.subplots(6, 1, figsize=(15,15))

for i,d in enumerate([df.query('platform == "PS4"')[relations].corr(), 
                      df.query('platform == "PS3"')[relations].corr(),
                      df.query('platform == "XOne"')[relations].corr(),
                      df.query('platform == "PSV"')[relations].corr(),
                      df.query('platform == "PC"')[relations].corr(),
                      df.query('platform == "X360"')[relations].corr(),]):

    sns.heatmap(d,
                cmap='twilight', 
                ax=ax[i],
                annot =True
               )
    ax[i].set_xlabel('Параметры модели')
    ax[i].set_ylabel('Параметры модели')

ax[0].set_title('Корреляция  продаж и рейтинга игр на  PS4')
ax[1].set_title('Корреляция  продаж и рейтинга игр на PS3')
ax[2].set_title('Корреляция  продаж и рейтинга игр на XOne')
ax[3].set_title('Корреляция  продаж и рейтинга игр на PSV')
ax[4].set_title('Корреляция  продаж и рейтинга игр на PC')
ax[5].set_title('Корреляция  продаж и рейтинга игр на X360')

plt.subplots_adjust(wspace= 1.5, hspace=1.5)
```

Как можем наблюдать из расчетов коэффициентов корреляции, подтверждается наше наблюдение о том, что связь между пользовательским рейтингом и продажами игры не наблюдается.  Между рейтингом критиков и продажами связь сильнее, но также ее нельзя назвать существенной. Большее влияние рейтинг критиков оказывает на продажи игр на новых платформах, а на продажи игр для ПК это влияние минимально из  отобранных нами платформ.

#### Проверка гипотез.

##### Средние  пользовательские рейтинги платформ Xbox One и PC.

Проверим отличаются ли рейтинги игр на разных платформах -возьмем популярную Xbox One и PC. Проведем t-test двух независимых выборок, поэтому используем метод ttest_ind с двусторонней альтернативной гипотезой, т.к. нас не интересует отличаются ли они в большую или меньшую сторону.
- H0 - рейтинги игр на платформах Xbox One и PC не отличаются;
- H1 - рейтинги игр не равны;

Проверим также сильно ли отличаются наши выборки по размеру

```python
len(df.query('platform == "XOne"')['user_score'].dropna())
```

```python
len(df.query('platform == "PC"')['user_score'].dropna())
```

Размеры выборок отличаются. Поэтому предположим, что дисперсия отличается, учтем это в расчетах.

```python
# выбираем уровень статистической значимости
alpha = 0.05


results = st.ttest_ind(df.query('platform == "XOne"')['user_score'].dropna(),
                       df.query('platform == "PC"')['user_score'].dropna(),
                       equal_var = False,
                       )
print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Пользовательские рейтинги игр отличаюся на платформе Xbox One и PC ')
else:
    print('Пользовательские рейтинги не отличаются')
```

Можно полагать, что средний пользовательский рейтинг не сильно отличается на платформе Xbox One и PC.

Посмотрим для подтверждения гипотезы на средние значения пользовательского рейтинга.

```python
'Среднее значение пользовательского рейтинга на Xbox One:\
{score1:.2f}, а среднее значение пользовательского рейтинга игр на ПК:\
{score2:.2f}'.format(score1 = df.query('platform == "XOne"')['user_score'].mean(),
                     score2 = df.query('platform == "PC"')['user_score'].mean())
```

Как мы видим, средние значения очень близки.

#####  Средние пользовательские рейтинги жанров Action и Sports.

Выдвинем гипотезу, что средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
Проведем t-test для двух независимых выборок, используем метод ttest_ind с двусторонней альтернативной гипотезой.
- H0 - пользовательские рейтинги жанров и Sports равны;
- H1 - рейтинги разные;

```python
len(df.query('genre == "action"')['user_score'].dropna())
```

```python
len(df.query('genre == "sports"')['user_score'].dropna())
```

Выборки разные по размеру, укажем , что дисперсия различна.

```python
# выбираем уровень статистической значимости
alpha = 0.05


results = st.ttest_ind(df.query('genre == "action"')['user_score'].dropna(),
                       df.query('genre == "sports"')['user_score'].dropna(),
                       equal_var = False,
                       )
print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Пользовательские рейтинги игр отличаются')
else:
    print('Нет оснований полагать, что пользовательские рейтинги отличаются')
```

А вот средний пользовательский рейтинг у игр в жанре экшен и спорт предположительно отличается. Посчитаем средние значения для выборок в подтверждение наших выводов.

```python
'Среднее значение пользовательского рейтинга игр в жанре экшен:\
{score1:.2f}, а среднее значение пользовательского рейтинга игр в жанре спорт:\
{score2:.2f}'.format(score1 = df.query('genre == "action"')['user_score'].mean(),
                     score2 = df.query('genre == "sports"')['user_score'].mean())
```

Как видим, средние действительно существенно отличаются.
